```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=[1,2,3,4]
y=[1,2,3,5]
plt.plot(x,y)
plt.show()
```

```python
x=[1,2,3,4]
y=[1,2,3,5]
plt.plot(x,y)
plt.title("Example Graph")
plt.show()
```

```python
x=[1,2,3,4]
y=[1,2,3,5]
plt.plot(x,y)
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
a=np.arange(1,10,0.5)
b=np.arange(1,10,0.5)
plt.plot(a,b)
plt.show()
```

```python
x=[1,2,3,4,5,6]
y=[1,2,3,4,5,6]
plt.plot(x,y,label="1:1",marker=".",linestyle="--",markersize="10",markeredgecolor="blue",color="green")
plt.show()
```
